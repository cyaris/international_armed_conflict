```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
```

```python
trade_df_group = pd.read_pickle('pickle/trade_df_group.pkl')
mat_cap_df = pd.read_pickle('pickle/mat_cap_df.pkl')
alliance_df_group = pd.read_pickle('pickle/alliance_df_group.pkl')
mid_df_group = pd.read_pickle('pickle/mid_df_group.pkl')
wrp_df = pd.read_pickle('pickle/wrp_df_imputed.pkl')
gov_df = pd.read_pickle('pickle/gov_df.pkl')
c_code_df = pd.read_pickle('pickle/c_code_df.pkl')
```

```python
one_df = pd.merge(trade_df_group, mat_cap_df, how = 'outer', on = ['c_code_1', 'year'])

one_df.rename({'state_name_1_x': 'state_name_1'}, axis = 1, inplace = True)

one_df.drop(['state_name_1_y'], axis = 1, inplace = True)
```

```python
c_code_dic = {}

for i, code in enumerate(c_code_df['c_code']):
    c_code_dic[code] = c_code_df['state_name'][i]
```

```python
for i, code in enumerate(one_df['c_code_1']):
    one_df.loc[i, 'state_name_1'] = c_code_dic[code]
```

```python
two_df = pd.merge(one_df, alliance_df_group, how = 'outer', on = ['c_code_1', 'year'])

two_df.rename({'state_name_1_x': 'state_name_1'}, axis = 1, inplace = True)

two_df.drop(['state_name_1_y'], axis = 1, inplace = True)
```

```python
for i, code in enumerate(two_df['c_code_1']):
    two_df.loc[i, 'state_name_1'] = c_code_dic[code]
```

```python
three_df = pd.merge(two_df, mid_df_group, how = 'outer', on = ['c_code_1', 'year'])

three_df.drop(['state_name_abb_a'], axis = 1, inplace = True)
```

```python
for i, code in enumerate(three_df['c_code_1']):
    three_df.loc[i, 'state_name_1'] = c_code_dic[code]
```

```python
four_df = pd.merge(three_df, wrp_df, how = 'outer', on = ['c_code_1', 'year'])

four_df.rename({'state_name_1_x': 'state_name_1'}, axis = 1, inplace = True)

four_df.drop(['state_name_1_y'], axis = 1, inplace = True)
```

```python
for i, code in enumerate(four_df['c_code_1']):
    four_df.loc[i, 'state_name_1'] = c_code_dic[code]
```

```python
df = pd.merge(four_df, gov_df, how = 'inner', on = ['c_code_1', 'year'])

df = df.sort_values(by = 'year', ascending = True).reset_index()

df.rename({'state_name_1_x': 'state_name', 'c_code_1': 'cow_code'}, axis = 1, inplace = True)

df.drop(['state_name_1_y', 'index'], axis = 1, inplace = True)
```

```python
for i, code in enumerate(df['cow_code']):
    df.loc[i, 'state_name'] = c_code_dic[code]
```
