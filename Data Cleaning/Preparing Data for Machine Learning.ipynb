```python
import warnings
from copy import deepcopy
import pandas as pd
import numpy as np
```

```python
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_colwidth', 1000)
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
np.random.seed(42)
```

```python
df = pd.read_pickle('../Pickles/df_imputed.pkl')
```

```python
majority_religion_df = pd.get_dummies(df['majority_religion'])
majority_religion_df.drop(['Christianity'], axis = 1, inplace = True)

un_region_df = pd.get_dummies(df['un_region'])
un_region_df.drop(['Western Africa'], axis = 1, inplace = True)

un_continent_df = pd.get_dummies(df['un_continent'])
un_continent_df.drop(['Africa'], axis = 1, inplace = True)

election_type_df = pd.get_dummies(df['election_type'])
election_type_df.drop(['Direct election'], axis = 1, inplace = True)

legislation_type_df = pd.get_dummies(df['legislation_type'])
legislation_type_df.drop(['Elective legislature'], axis = 1, inplace = True)

legislature_status_df = pd.get_dummies(df['legislature_status'])
legislature_status_df.drop(['Unknown'], axis = 1, inplace = True)

party_legal_status_df = pd.get_dummies(df['party_legal_status'])
party_legal_status_df.drop(['Unknown'], axis = 1, inplace = True)

party_existance_df = pd.get_dummies(df['party_existance'])
party_existance_df.drop(['Unknown'], axis = 1, inplace = True)

party_existance_outside_regime_df = pd.get_dummies(df['party_existance_outside_regime'])
party_existance_outside_regime_df.drop(['Unknown'], axis = 1, inplace = True)

legislature_parties_df = pd.get_dummies(df['legislature_parties'])
legislature_parties_df.drop(['Unknown'], axis = 1, inplace = True)

incumbent_type_df = pd.get_dummies(df['incumbent_type'])
incumbent_type_df.drop(['Democratic incumbent'], axis = 1, inplace = True)

collective_leadership_df = pd.get_dummies(df['collective_leadership'])
collective_leadership_df.drop(['Unknown'], axis = 1, inplace = True)
collective_leadership_df.drop(['No'], axis = 1, inplace = True)
collective_leadership_df.columns = ['collective_leadership']

military_leader_df = pd.get_dummies(df['military_leader'])
military_leader_df.drop(['No'], axis = 1, inplace = True)
military_leader_df.columns = ['military_leader']

royal_leader_df = pd.get_dummies(df['royal_leader'])
royal_leader_df.drop(['No'], axis = 1, inplace = True)
royal_leader_df.columns = ['royal_leader']

nominal_vs_eff_diff_df = pd.get_dummies(df['nominal_vs_eff_diff'])
nominal_vs_eff_diff_df.drop(['No'], axis = 1, inplace = True)
nominal_vs_eff_diff_df.columns = ['nominal_vs_eff_diff']

communist_leader_df = pd.get_dummies(df['communist_leader'])
communist_leader_df.drop(['No'], axis = 1, inplace = True)
communist_leader_df.columns = ['communist_leader']

leader_died_df = pd.get_dummies(df['leader_died'])
leader_died_df.drop(['No'], axis = 1, inplace = True)
leader_died_df.columns = ['leader_died']

democratic_regime_df = pd.get_dummies(df['democratic_regime'])
democratic_regime_df.drop(['No'], axis = 1, inplace = True)
democratic_regime_df.columns = ['democratic_regime']

cabinet_assembly_df = pd.get_dummies(df['cabinet_assembly'])
cabinet_assembly_df.drop(['No'], axis = 1, inplace = True)
cabinet_assembly_df.columns = ['cabinet_assembly']

popular_election_df = pd.get_dummies(df['popular_election'])
popular_election_df.drop(['No'], axis = 1, inplace = True)
popular_election_df.columns = ['popular_election']

regime_type_df = pd.get_dummies(df['regime_type'])

transition_to_democracy_df = pd.get_dummies(df['transition_to_democracy'])
transition_to_democracy_df.drop(['No'], axis = 1, inplace = True)
transition_to_democracy_df.columns = ['transition_to_democracy']

transition_to_dictatorship_df = pd.get_dummies(df['transition_to_dictatorship'])
transition_to_dictatorship_df.drop(['No'], axis = 1, inplace = True)
transition_to_dictatorship_df.columns = ['transition_to_dictatorship']

war_vs_peace_df = pd.get_dummies(df['war_vs_peace'])
war_vs_peace_df.drop(['peace'], axis = 1, inplace = True)
war_vs_peace_df.columns = ['war_present']

conflict_vs_peace_df = pd.get_dummies(df['conflict_vs_peace'])
conflict_vs_peace_df.drop(['peace'], axis = 1, inplace = True)
conflict_vs_peace_df.columns = ['conflict_present']

decade_df = pd.get_dummies(df['decade'])
decade_df.drop(['2000s'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.drop(['majority_religion', 'un_region', 'un_continent', 'election_type',
         'legislation_type', 'legislature_status', 'party_legal_status',
         'legislation_type', 'party_existance', 'party_existance_outside_regime',
         'legislature_parties', 'incumbent_type', 'collective_leadership',
         'military_leader', 'royal_leader', 'nominal_vs_eff_diff',
         'communist_leader', 'democratic_regime', 'cabinet_assembly',
         'popular_election', 'regime_type', 'transition_to_democracy',
         'transition_to_dictatorship', 'leader_died', 'war_vs_peace',
         'conflict_vs_peace', 'decade'], axis = 1, inplace = True)

df = deepcopy(pd.concat([df, un_continent_df, election_type_df, legislation_type_df,
                         legislature_status_df, party_legal_status_df,
                         party_existance_df, party_existance_outside_regime_df,
                         legislature_parties_df, incumbent_type_df,
                         collective_leadership_df, military_leader_df,
                         royal_leader_df, nominal_vs_eff_diff_df,
                         communist_leader_df, leader_died_df,
                         democratic_regime_df, cabinet_assembly_df,
                         popular_election_df, regime_type_df,
                         transition_to_democracy_df, transition_to_dictatorship_df,
                         war_vs_peace_df, conflict_vs_peace_df, decade_df],
                        axis = 1, join = 'outer'))
```
